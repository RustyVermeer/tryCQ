## Totally new to Jupyter?

Don't worry, the basics are nice and simple to grasp :). 

Jupyter's main page provides a file browser, allowing you to select folders and files. The most interesting of which are *notebook* files with the file extension **.ipynb**. You're interacting with one such file now, in fact.

Inside a notebook you have a view of *cells* which can contain either text or code. This cell is a *markdown* cell used for inline explanations and documentation. The cell below is a *code* cell, which is interpreted by the **kernel** as python code. If you ever set up your own installation, you will have the ability to install different kernels that allow you to create notebooks for different languages. Powerful stuff!

You can use the menus at the top of the notebook to add and change cells. Play around a bit and see what you can do! 

Feel free to open new cells using the plus button (<button class='fa fa-plus icon-plus btn btn-xs btn-default'></button>), or hitting shift-enter while this cell is selected.

Behind the scenes, the software that powers this is [tmpnb](https://github.com/jupyter/tmpnb), a  Tornado application that spawns [pre-built Docker containers](https://github.com/ipython/docker-notebook) and then uses the [jupyter/configurable-http-proxy](https://github.com/jupyter/configurable-http-proxy) to put your notebook server on a unique path.

## Totally new to CadQuery?

That's cool, too. It's a good thing you're here, then! 

Have a look through the examples folder to get a hang of the library.
In order to get a nice display of the geometry, you'll have to follow this simple sequence of steps:

1. Highlight the cell containing the CadQuery (often shortened to CQ) script. 
2. Press CTRL + ENTER or hit the RUN button.
3. Once the cell is run, you'll see a few boxes input boxes that appear alongside a *Run Interact* button. Click the button.
4. Have a look at your Geometry! Scroll wheel will zoom in and out. Hold Left click and drag to rotate the object around. Hold Right Click and drag to shift the frame.
5. Change values around in the inputs and press *Run Interact* again to regenerate a new geometry. Re-run the cell to reset everything.
